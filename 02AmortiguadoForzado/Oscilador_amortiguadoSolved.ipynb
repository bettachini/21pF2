Para simplificar la formula llamé:


*   $a=\frac{-\Gamma}{2}$
*   $b=\sqrt{w_0^2-\frac{-\Gamma^2}{4}}$

Condiciones Iniciales:


*   $X_0=x_0$
*   $\dot X_0=\dot x_0$

Formula general de movimiento:

$X_h(t) = \frac{x_0}{cos(arctan\frac{x_0.a-\dot x_0}{x_0.b})} e^{a.t}.cos[b.t + (arctan\frac{x_0.a-\dot x_0}{x_0.b})]$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab, mlab, pyplot
import ipywidgets as iwy
```

```python
def function(omega_0 = 0.1, gamma = 0.1, x = 0.1, v = 0.1, tiempo= 50):
    #Datos de la ecuación
    a= -gamma/2
    b= omega_0**2+(gamma**2/4)
    c= np.sqrt(b)
    d= (x**2)*(c**2)+(x**2)*(a**2)-2*a*v*x+(v**2)

    theta= (np.arctan((x*a-v)/x*c))
    t= np.linspace(0,tiempo,250)
    ec1=(x/np.cos(theta))*np.e**(a*t)*np.cos(c*t+theta) #Ecuación de movimiento

    #Grafico de la función
    #En esta sección escribo las 3 posibilidades del oscilamiento amortiguado
    if omega_0**2 > (gamma*2)/4:
        plt.plot(t, ec1, color= "r", label="Caso Subamortiguado");
    elif omega_0**2 < (gamma*2)/4:
        plt.plot(t, ec1, color= "r", label="Caso Sobreamortiguado");
    else:
        plt.plot(t, ec1, color= "r", label="Caso Crítico");
    plt.suptitle('Oscilador Amortiguado', fontsize=14, y=1)
    plt.xlabel('tiempo [s]', fontsize=12)
    plt.ylabel('Oscilación', fontsize=12)
    plt.grid()
    plt.legend()
```

```python
iwy.interact(function,omega_0=(0.1,1,0.01),gamma=(0.01,1,0.01),x=(0.1,20,0.01),v=(0.1,10,0.01),tiempo=(0,250,1));
```
