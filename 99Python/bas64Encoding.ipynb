## How to encode

```python
import base64
```

```python
# converts to base64
# img_file = 'malla.png'
img_file = 'logoDF.png'
encoded = base64.b64encode(open(img_file, "rb").read()) # esto lo engloba entre b'<contenido>' 
pasteThis = '!['+ img_file+ '](data:image/png;base64,{}'.format(str(encoded)[2:-1]) + ')'
```

```python
# write file
with open(img_file[:-3]+'txt', 'w') as f:
    f.write(pasteThis)
```
