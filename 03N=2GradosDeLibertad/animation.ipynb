# Jouer à être un animateur

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from IPython.display import HTML
mpl.rc('animation', html='html5')
```

```python
# Este es felizmente tan conciso.
t = np.linspace(0,2*np.pi)
x = np.sin(t) # pre-calcula, eso es bueno

fig, ax = plt.subplots()
ax.axis([0,2*np.pi,-1,1])
l, = ax.plot([],[])

def animate(i):
    l.set_data(t[:i], x[:i]);

ani = mpl.animation.FuncAnimation(fig, animate, frames= len(t) );
```

```python
# Tomado de página
# https://matplotlib.org/stable/api/animation_api.html

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r.')

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = mpl.animation.FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'ro')

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
```

```python
# plt.style.use('seaborn-pastel')
fig = plt.figure()
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))
line, = ax.plot([], [], lw=3)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = np.linspace(0, 4, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,

anim = mpl.animation.FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)
```
