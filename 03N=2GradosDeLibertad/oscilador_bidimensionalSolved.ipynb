```python
import sympy as sp
from sympy.physics.mechanics import init_vprinting
init_vprinting()
```

```python
k,a,m=sp.symbols("K a M", positive=True)
la,lb,lc,ld=sp.symbols("L_A L_B L_C L_D",positive=True)
alpha,beta,gamma,delta=sp.symbols("alpha beta gamma delta",positive=True)
t=sp.symbols("t")
x=sp.Function("x")(t)
y=sp.Function("y")(t)
```

```python
newton=[sp.Eq(sp.diff(x,t,2),(k/m)*(-la*sp.cos(alpha)+lb*sp.cos(beta)-lc*sp.sin(gamma)-ld*sp.sin(delta))),
       sp.Eq(sp.diff(y,t,2),(k/m)*(-la*sp.sin(alpha)-lb*sp.sin(beta)+lc*sp.cos(gamma)-ld*sp.cos(delta)))]
newton
```

```python
cosenos=[sp.Eq(sp.cos(alpha),(a+x)/la),
        sp.Eq(sp.cos(beta),(a-x)/lb),
        sp.Eq(sp.cos(gamma),(a-y)/lc),
        sp.Eq(sp.cos(delta),(a+y)/ld)]
cosenos
```

```python
senos=[sp.Eq(sp.sin(alpha),y/la),
        sp.Eq(sp.sin(beta),y/lb),
        sp.Eq(sp.sin(gamma),x/lc),
        sp.Eq(sp.sin(delta),x/ld)]
senos
```

```python
newton_subs=[]
for i in range(2):
    newton_subs.append(newton[i].subs([(sp.cos(alpha),cosenos[0].rhs),
                        (sp.cos(beta),cosenos[1].rhs),
                        (sp.cos(gamma),cosenos[2].rhs),
                        (sp.cos(delta),cosenos[3].rhs),
                        (sp.sin(alpha),senos[0].rhs),
                        (sp.sin(beta),senos[1].rhs),
                        (sp.sin(gamma),senos[2].rhs),
                        (sp.sin(delta),senos[3].rhs)]))
newton_subs
```

```python
sp.Eq(sp.Matrix([x,y]),sp.Matrix([sp.dsolve(newton_subs[0]).rhs,sp.dsolve(newton_subs[1]).rhs]))
```
