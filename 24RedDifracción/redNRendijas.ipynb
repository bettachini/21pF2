## Enunciado
Una onda plana monocromática de longitud de onda \\(\lambda\\) incide normalmente sobre una red de transmisión plana formada por \\(N\\) rendijas de ancho \\(b\\) y período \\(d\\).
Suponiendo que la teoría corresponde a una descripción exacta del fenómeno:   
1. Analice la distribución de intensidad sobre la pantalla y grafíquela.  
2. Calcule:   
    1. La posición angular de las líneas espectrales (¿a qué máximos corresponden?), y su intensidad.   
    2. El número de mínimos de interferencia entre dos líneas espectrales, por ende, ¿cuántos máximos secundarios hay?   
    3. El ancho angular de las líneas espectrales.   
    4. El máximo orden observable.   
3. Discuta:   
    1. ¿Qué aproximación hace en los ángulos?    
    2. La dependencia de los parámetros con el número de rendijas y con la densidad de rendijas.   

## Resolución en la aproximación de campo lejano
Cada punto dentro de las aperturas de ancho \\(b\\) distribuidas períodicamente en una distancia \\(d\\) los consideramos emisores puntuales de **frentes de ondas esféricas** que se propagan en un medio en dirección \\(\vec{k}\\).
Usualmente nos sirve dibujar las direcciónes \\(\vec{k}\\) con  **rayos** para que sea visualmente clara la dirección en que se están propagando dichas ondas.

![](figuras/redDifracciónTransmisión.png)

Esta ilustración asume la **aproximación de campo lejano**
- Red de difracción y pantalla enfrentadas a una distancia suficiente \\(D\\).
- Desde el centro de la red a un \\(x\\) en la pantalla hay una \\(R \simeq D\\).
- Desde cualquier punto emisor en las rendijas hasta \\(x\\) hay una distancia \\(r \simeq R \simeq D\\).

El campo que se percibirá en un punto de la pantalla \\(x\\) emitido desde uno de estos recorriendo una distancia \\(r\\).
$$
\vec{E}_1(x,t) = \frac{E_0}{r} e^{i(\vec{k} \vec{r} - \omega t ) } \hat{y},
$$
siendo \\(\hat{y}\\) perpendicular a \\(\vec{k}\\).

### Campo eléctrico en pantalla por un solo emisor
La diferencia de camino geométrico entre \\(r = r(x')\\) y \\(R\\) medida desde el centro de la red es la misma que en el interferómetro de Young.
$$
r \simeq R - x' \sin(\theta),
$$
donde \\(\theta\\) es el ángulo desde el centro de la red hasta el punto en la pantalla \\(x\\),

Así 
$$
\vec{E}_1(x,t) \simeq \frac{E_0}{R} e^{i (k R - \omega t)} e^{-i k x' \sin(\theta) } \hat{y},
$$
donde se hace caso omiso de la reducción de la amplitud por el alejamiento de la fuente amparandose en la aproximacíon \\(r\simeq R\\).

#### Campo eléctrico en pantalla por todos los emisores en las rendijas
Para considerar el continuo de emisores se integra a lo largo de la extensión de las rendijas.
Estas van en \\( d(m+\frac{1}{2}) - \frac{b}{2} < x' < d(m+\frac{1}{2}) + \frac{b}{2} \\) siendo que \\(-\frac{N}{2} < m < \frac{N}{2}-1\\) con \\(m \in \mathbb{Z} \\), así 
$$
E(x,t) \simeq \frac{E_0 e^{i (k R - \omega t)}}{R} 
  \sum_{m = -\frac{N}{2}}^{\frac{N}{2} - 1} \int_{ d(m+\frac{1}{2}) - \frac{b}{2} }^{ d(m+\frac{1}{2}) + \frac{b}{2} }  e^{-i k x' \sin(\theta) } d x' .
$$

Estos complicados límites de integración "claman" por un cambio de varible.
Aplicando \\(x" = x' - d (m + \frac{1}{2}) \\) permite integrar en la extensión de una única rendija
$$
\begin{aligned}
E(x,t)
&\simeq \frac{E_0 e^{i (k R - \omega t)}}{R}
  \sum_{m = -\frac{N}{2}}^{\frac{N}{2} - 1 } \int_{- \frac{b}{2} }^{\frac{b}{2} }  e^{-i k \left[ x" + d (m + \frac{1}{2}) \right] \sin(\theta) } d x" \\
&\simeq \frac{E_0 e^{i (k R - \omega t)}}{R} e^{- i k \frac{d}{2} \sin(\theta) }
  \sum_{m = -\frac{N}{2}}^{\frac{N}{2} - 1 } e^{-i k m d \sin(\theta) } \int_{- \frac{b}{2} }^{\frac{b}{2} }  e^{-i k x'' \sin(\theta) } d x'' 
\end{aligned}
$$

Para simplificar la notación lo que permanece constante entre rendijas se resume en \\(\tilde{E}_0\\).
Se resuelve la integral en una rendija
$$
\begin{aligned}
E(x,t)
&\simeq \tilde{E}_0
  \sum_{m = -\frac{N}{2}}^{\frac{N}{2} - 1 } e^{-i k m d \sin(\theta) } \left[ \frac{ e^{-i k x'' \sin(\theta) } }{ -i k \sin(\theta) } \right]_{x'' = - \frac{b}{2} }^{x'' = \frac{b}{2} }
 = \tilde{E}_0 
 \sum_{m = -\frac{N}{2}}^{\frac{N}{2} - 1 } e^{-i k m d \sin(\theta) } \left[ \frac{ e^{-i k \frac{b}{2} \sin(\theta) } - e^{i k \frac{b}{2} \sin(\theta) } }{ -i k \sin(\theta) } \right] \\
&= \tilde{E}_0 
 \sum_{m = -\frac{N}{2}}^{\frac{N}{2}- 1} e^{-i k m d \sin(\theta) } \left[ \frac{\frac{b}{2}}{\frac{b}{2}} \frac{ -2 i \sin{ \left(k \frac{b}{2} \sin(\theta) \right) } }{ -i k \sin(\theta) } \right] 
 = \tilde{E}_0 
 \sum_{m = -\frac{N}{2}}^{\frac{N}{2} - 1} e^{-i k m d \sin(\theta) } \left[ b\, \mathrm{sinc} \left( k \frac{b}{2} \sin(\theta) \right) \right] 
\end{aligned}
$$
donde lo que permanecía constante entre rendijas se resumió en \\(\tilde{E}_0\\).

Usualmente el argumento del \\(\mathrm{sinc}\\) se suele resumir en una letra 
$$
    \beta = k \frac{b}{2} \sin(\theta) = \frac{2 \pi n}{\lambda_0} \frac{b}{2} \sin(\theta),
$$
donde \\(n\\) es el índice de refracción del medio en que las ondas se propagan, con lo que se resume la contribución en \\(x\\)
$$
    E(x,t) \simeq \tilde{E}_0 \sum_{m = -\frac{N}{2}}^{\frac{N}{2} - 1 } e^{-i k m d \sin(\theta) } \left[ b\, \mathrm{sinc} \left( \beta \right) \right]
$$

De manera similar se hace práctico resumir 
$$
    \gamma = k d \sin(\theta) = \frac{2\ \pi n}{\lambda_0} d \sin(\theta),
$$
con lo que obtenemos una expresión más compacta
$$
    E(x,t) \simeq \tilde{E}_0 \sum_{m = -\frac{N}{2}}^{\frac{N}{2} - 1} e^{-i \gamma m } \left[ b\, \mathrm{sinc} \left( \beta \right) \right]
$$

Para llevar la sumatoria a la forma de una geómetrica
$$
  \sum_{i=0}^{N} r^i = \frac{r^{N+ 1} - 1}{r - 1},  
$$
debe salvarse que su suma se extiende desde \\(m\\) negativos.
Basta con un cambio de variable sobre la que se suma para esto \\( p = m + \frac{N}{2} \\)
$$
E(x,t)
\simeq \tilde{E}_0 \sum_{p = 0}^{N-1} e^{-i \gamma ( p- \frac{N}{2} ) } \left[ b\, \mathrm{sinc} \left( \beta \right) \right] 
= \tilde{E}_0 e^{i \gamma \frac{N}{2} } \sum_{p = 0}^{N - 1} e^{-i \gamma p } \left[ b\, \mathrm{sinc} \left( \beta \right) \right].
$$

Se utiliza ahora el resultado de la sumatoria
$$
\begin{aligned}
E(x,t)
&\simeq \tilde{E}_0 e^{i \gamma \frac{N}{2} } \sum_{p = 0}^{N-1} \left( e^{-i \gamma } \right)^p \left[ b\, \mathrm{sinc} \left( \beta \right) \right]
= \tilde{E}_0 e^{i \gamma \frac{N}{2} } \left[ \frac{ e^{-i \gamma N} - 1 }{ e^{-i \gamma} - 1 } \right] \left[ b\, \mathrm{sinc} \left( \beta \right) \right] \\
&\simeq \tilde{E}_0 e^{i \gamma \frac{N}{2} } \left[ \frac{ e^{-i \gamma \frac{N}{2} } }{ e^{-i \frac{\gamma}{2} } } \right] \left[ \frac{ e^{-i \gamma \frac{N}{2} } - e^{i \gamma \frac{N}{2} } }{ e^{-i \frac{\gamma}{2} } - e^{i \frac{\gamma}{2} } } \right] \left[ b\, \mathrm{sinc} \left( \beta \right) \right]
 = \tilde{E}_0 e^{i \frac{\gamma}{2} } \left[ \frac{ -2 i \sin( \gamma \frac{N}{2} ) }{ -2 i \sin( \frac{\gamma}{2} ) } \right] \left[ b\, \mathrm{sinc} \left( \beta \right) \right] \\
&\simeq \tilde{E}_0 e^{i \frac{\gamma}{2} } \left[ \frac{ \sin( \frac{\gamma}{2} N ) }{ \sin( \frac{\gamma}{2} ) } \right] \left[ b\, \mathrm{sinc} \left( \beta \right) \right]
\end{aligned}
$$

### Calculo de la irradiancia
La expresión de irradiancia es
$$
I(x) = \frac{n c \varepsilon_0}{2} \langle \vec{E}(x,t) \rangle^2_\tau = \frac{n c \varepsilon_0}{2} \frac{1}{\tau} \int_t^{t + \tau} \left( \vec{E}(x,t') \right)^2 dt' ,
$$
donde figura la velocidad de la luz \\(c\\) y la permitividad eléctrica en el vacio \\( \varepsilon_0 \\), junto a \\(n\\), el índice de refracción del medio.

$$
\begin{aligned}
I(x) 
& \simeq \frac{n c \varepsilon_0}{2} E(x,t) E^*(x,t) = 
\frac{n c \varepsilon_0}{2} \tilde{E}^2_0 e^{-i \frac{\gamma}{2} } e^{i \frac{\gamma}{2} }
\left[ \frac{ \sin^2( \frac{\gamma}{2} N ) }{ \sin^2( \frac{\gamma}{2} ) } \right] \left[ b^2 \, \mathrm{sinc}^2 \left( \beta \right) \right] \\
& \simeq I_0 \frac{ \sin^2( \frac{\gamma}{2} N ) }{ \sin^2( \frac{\gamma}{2} ) } \mathrm{sinc}^2 \left( \beta \right)
\end{aligned}
$$

## Graficación de la distribución de intensidad sobre la pantalla

```python
import numpy as np # biblioteca de cálculo numérico
```

```python
# Iluminación
I_0 = 100* 1E2**2 # [W m-2]
n_p = 1 # del medio en la pantalla
# E0 = np.sqrt(2* I_0/ (constants.epsilon_0* constants.c* n_p) ) # [V m-1]
lambda_0 = 400E-9 # [m] azul
colorMap = 'Blues'
```

```python
def interferenciaN(n_p, lambda_0, d, N, sinsTheta):
    '''
    sinsTheta : senos de los ángulos con que incide en un x' en la pantalla
    '''
    gamma = (2* np.pi* n_p/ lambda_0)* d* sinsTheta
    return (np.sin(gamma* N/2)/ np.sin(gamma/2) )**2

def difraccion1(n_p, lambda_0, b, sinsTheta):
    beta = (2* np.pi* n_p/ lambda_0)* (b/2)* sinsTheta
    return (np.sin(beta)/ beta)**2
```

```python
# Parámetros de red de rendijas
N = 10 # número de rendijas
b = 50E-6 # [m]
d = 100E-6 # [m]

# Parametros del sistema de observación
D = 2 # [m] distancia red a pantalla
x_semi = 6E-2 # [m] semi-ancho de la pantalla de observación
```

```python
# biblioteca de graficación
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
## Graficación convencional
sinExtremoPantalla = x_semi/np.sqrt(x_semi**2 + D**2) # [m]
puntos = np.int(1E3)
sinsTheta = np.linspace(-sinExtremoPantalla, sinExtremoPantalla, num = puntos)
xPrimaPantalla = D* np.tan(np.arcsin( sinsTheta ) )  # [m] posiciones en la pantalla
```

```python
# Calculo de irradiancia en cada punto
interferencia = interferenciaN(n_p, lambda_0, d, N, sinsTheta)
difraccion = difraccion1(n_p, lambda_0, b, sinsTheta)
```

```python
# posición en pantalla
fig, ax = plt.subplots(3,1,figsize=(16,12), sharex=True)
fig.suptitle('Irradiancia en pantalla')
I_interferencia = interferencia * I_0* 1E-2**2 # [W cm^-2]
ax[0].plot(xPrimaPantalla * 1e3, I_interferencia)
#ax[0].set_xlabel('x [mm]')
ax[0].set_ylabel(r'Irradiancia [$\mathrm{\frac{W}{cm^2}}$]')
I_difraccion = difraccion* I_0* 1E-2**2 # [W cm^-2]
ax[1].plot(xPrimaPantalla * 1e3, I_difraccion)
# ax[1].set_xlabel('x [mm]')
ax[1].set_ylabel(r'Irradiancia [$\mathrm{\frac{W}{cm^2}}$]')
I_total = interferencia* difraccion* I_0* 1E-2**2 # [W cm^-2]
ax[2].plot(xPrimaPantalla * 1e3, I_total )
ax[2].set_xlabel('x [mm]')
ax[2].set_ylabel(r'Irradiancia [$\mathrm{\frac{W}{cm^2}}$]')
```

## Calcule:
### La posición angular de las líneas espectrales (¿a qué máximos corresponden?), y su intensidad.

Como puede verse en la graficación las **lineas espectrales** son los máximos principales de interferencia.
El término de interferencia
$$
\frac{ \sin^2( \frac{\gamma}{2} N ) }{ \sin^2( \frac{\gamma}{2} ) }
$$
se máximiza allí donde su numerador tienda a anularse.
Esto es cuando 
$$
\sin{\frac{\gamma}{2} } = 0 \implies \frac{\gamma}{2} = M \pi\,(M \in \mathbb{N}_0)
$$
$$
\frac{\pi n}{\lambda_0} d \sin(\theta_m) = M \pi
$$
$$
\sin(\theta_M) = M \frac{\lambda_0}{d n}
$$
Uno en particular que nos permite calcular en forma sencilla la intensificación de la intensidad, es cuando \\(\gamma \to 0\\)
$$
\lim_{\gamma \to 0} \frac{ \sin^2( \frac{\gamma}{2} N ) }{ \sin^2( \frac{\gamma}{2} ) } = N^2,
$$
por lo que **la irradiancia se intensifica con el cuadrado del número de rendijas**.

### El número de mínimos de interferencia entre dos líneas espectrales, por ende, ¿cuántos máximos secundarios hay?
El termino de interferencia se minimiza para cuando el numerador se anula, es decir
$$
\sin{\frac{\gamma}{2} N} = 0 \implies \frac{\gamma}{2} N = m \pi\,(m \in \mathbb{N}_0)
$$
$$
\frac{\pi n}{\lambda_0} d \sin(\theta_m) N = m \pi
$$
$$
\sin(\theta_m) = m \frac{\lambda_0}{d n N}.
$$
Vemos que estos son una subdivisión de la frecuencia de los \\(\sin(\theta_M)\\) dividida por el número de rendijas \\(N\\).
Entre estos \\(m=0\\) y \\(m=N\\) correspondería a los \\(\sin(\theta_M)\\), pero los \\(m=1,2,\dots,N-1 \\) son \\( \sin(\theta_m) \\). Entonces **hay \\(N-1\\) mínimos de interferencia entre máximos**.

### El ancho angular de las líneas espectrales.
Este ancho por cualquier criterio que se utilice (e.g. ancho a media altura) será menor o igual al que hay entre \\(\sin(\theta_m)\\) a cada lado del máximo de interferencia.
Tomemos los adyantel al central central
$$
\Delta \sin(\theta_m) = \sin(\theta_{m=1}) - \sin(\theta_{m=-1}) = \frac{\lambda_0}{d n N} \left(1 - (-1) \right) = \frac{2\lambda_0}{d n N}.
$$

### El máximo orden observable. 
Viendo el gráfico de la irradiancia no es difícil convencerse de que los únicos máximos de interfrencia son aquellos que se alojan dentro de la campana principal de difracción.
Los límites de cada campana corresponden a cuando en el término de difracción
$$
\mathrm{sinc}^2 \left( \beta \right) = \left( \frac{\sin{\beta}}{\beta} \right)^2,
$$
se anula el numerador, es decir
$$
\sin{\beta} = 0 \implies \beta = q \pi \, (p \in \mathbb{N}_0)
$$
$$
\frac{2 \pi n}{\lambda_0} \frac{b}{2} \sin(\theta_p) = q \pi
$$
$$
\sin(\theta_p) = q \frac{\lambda_0}{b n}.
$$
En el particular de la primer campana los límites serán para \\(q = \pm 1\\), por lo que debe cumplirse
$$
\left| \sin(\theta_{M}) \right| < \sin(\theta_{q=1})
$$
$$
-\frac{\lambda_0}{b n} < M \frac{\lambda_0}{d n} < \frac{\lambda_0}{b n}
$$
$$
-\frac{d}{b} < M < \frac{d}{b}
$$
