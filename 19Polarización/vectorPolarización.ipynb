**Polarización**

Animación de una onda transversal $\psi(z,t)$ vista en el plano $xy$. En las primeras lineas del código dejo las variables que se pueden modificar para variar el estado de polarización (amplitudes y fase)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML
rc('animation', html='html5')

w = 5 #frec
pi = np.pi

### PARÁMETROS ###

Ex = 1      #Amplitud X
Ey = 2      #Amplitud Y
PHI = pi/4  #Diferencia de fase

### PARÁMETROS ###

plt.style.use('bmh')

fig, ax = plt.subplots(figsize=(7,7))
plt.close()
ax.axis([-3,3,-3,3])
ax.set_aspect("equal")
ax.set_xlabel('$X$')
ax.set_ylabel('$Y$')

n = int(PHI/pi) if float(PHI/pi).is_integer() else np.round(PHI/pi,2)
ax.set_title(f'$Polarización$ \n $E_x = {Ex} \\ \\ \\ E_y = {Ey} \\ \\ \\ \Delta\phi = {n}\pi$', fontsize=18, y=1.02) #titulo

psi = plt.Arrow(0,0,0,0)
ax.add_patch(psi)
line, = ax.plot([],[], color=plt.cm.viridis(.6), alpha=.3) #trazado

T = 2*pi/w #período
t = np.linspace(0,5*T,200)
psi_x = Ex*np.cos(-w*t)
psi_y = Ey*np.cos(-w*t + PHI)

def animacion(i): 
    line.set_data(psi_x[:i], psi_y[:i])
    ax.patches.pop(0) 
    psi = plt.Arrow(0, 0, psi_x[i], psi_y[i], width=.3, color=plt.cm.magma(.6), zorder=3)
    ax.add_patch(psi)
    return line, psi

anim = animation.FuncAnimation(fig, animacion, interval=30, frames=200)
anim
```
