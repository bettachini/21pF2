## La precesión de Larmor
![](https://upload.wikimedia.org/wikipedia/commons/thumb/6/6c/Precession_in_magnetic_field.svg/96px-Precession_in_magnetic_field.svg.png)  
Fuente: https://en.wikipedia.org/wiki/Larmor_precession  
En negro el momento dipolar magnético de la partícula, \\(\vec{\mu}\\), y en línea de puntos la rotación que causa el torque, \\(\vec{\tau}\\), que le produce un campo magnético, \\(\vec{B}\\), en verde.  

\\[
    \vec{\tau} = - \vec{\mu} \times \vec{B} = - \frac{e g}{2 m} \vec{J} \times \vec{B},
\\]
donde en el momento angular
\\[
    \vec{J} = \vec{L} + \vec{S},
\\]
se combinan el momento angular orbital, \\(\vec{L}\\), y el spin, \\(\vec{S}\\).

Las [cuentas intermedias de tal interacción](https://depts.washington.edu/jrphys/ph315W08/spinmag.pdf) involucran conceptos de electromagnetísmo y cuántica.
Aquí las obviamos.
Lo interesante es que se resuelven en una ecuación lineal de primer orden cuya solución es una oscilación armónica con
\\[
    \omega_\mathrm{Larmor} = - \frac{e g}{2 m} B.
\\]

Midiendo \\(\omega_\mathrm{Larmor}\\) en un \\(\vec{B}\\) conocido se determina la [razón o tasa giromagnética](https://es.wikipedia.org/wiki/Relaci%C3%B3n_giromagn%C3%A9tica) que contiene el [factor g](https://es.wikipedia.org/wiki/Factor-g_(f%C3%ADsica)).

  
El número de positrones registrados en un calorímetro del experimento oscila según \\(\omega_\mathrm{Larmor}\\).

## El factor \\(g\\)
Ese [factor g](https://es.wikipedia.org/wiki/Factor-g_(f%C3%ADsica)) se puede hacer adimensional
\\[
    \vec{\mu} = g \frac{\mu_B}{\hbar} \vec{J},
\\]
expresando  el momento magnético \\(\mu\\) en función del [magnetón de Bohr](https://es.wikipedia.org/wiki/Magnet%C3%B3n_de_Bohr), \\(\mu_B\\).

Con la (mayúsculas) [Ecuación de Dirac](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_de_Dirac), que aplica solo primeros principios de la cuántica, se calcula para la interacción con el spin del electrón \\(g = 2\\) (nada de \\(\vec{L}\\)) .

Y eso daban las mediciones en los años 20.
Pero en los 40 se midió una [anomalía en su momento magnético](https://en.wikipedia.org/wiki/Anomalous_magnetic_dipole_moment), es decir un \\(g \neq 2\\) más allá del error experimental.

En 1948 se calculó el efecto de interacciones con fotónes virtuales que se crean espontáneamente y vuelven a re-absorberse en
\\[
    a_e = \frac{g -2}{2} \simeq 0.001 161 4 ,
\\]
explicando el resultado experimental.

Esta historia se resume en la recomendable revista en línea conjunta del Fermilab y el SLAC de Stanford: https://www.symmetrymagazine.org/article/the-mystery-of-the-muons-magnetism

## Momento magnético anómalo del muón
Con el tiempo se mejoró el cálculo contemplando interacciones con
- fotones virtuales en el marco de la [electrodinámica cuántica (QED, en inglés)](https://es.wikipedia.org/wiki/Electrodin%C3%A1mica_cu%C3%A1ntica),
- [hadrones](https://es.wikipedia.org/wiki/Hadr%C3%B3n) virtuales,
- y bosones de la [fuerza electrodébil](https://es.wikipedia.org/wiki/Modelo_electrod%C3%A9bil) (EW, en inglés).

En el marco del [modelo estándar de las partículas](https://es.wikipedia.org/wiki/Modelo_est%C3%A1ndar_de_la_f%C3%ADsica_de_part%C3%ADculas) (SM, en inglés) para el **muón**, un pariente más masivo del electrón pero de igual spin
$$
a_\mu^{SM} = a_\mu^{QED} + a_\mu^{EW} + a_\mu^\mathrm{Hadrón} = 0.00116591810(43)
$$

El problema es que de las mediciones surge que
\\[
    a_\mu^\mathrm{experimental} = 0.00116592061(41) \neq a_\mu^{SM}
\\]

  
Fuente: Measurement of the Positive Muon Anomalous Magnetic Moment to 0.46 ppm
B. Abi et al. (Muon g−2 Collaboration)  
Phys. Rev. Lett. 126, 141801 – Published 7 April 2021
https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.126.141801
