```python
import numpy as np
import matplotlib.pyplot as plt
```

# SOLUCIÓN ANALÍTICA

$$ \theta (t) = A\sin(\sqrt{\frac{g}{L}}t + \phi)$$ 


```python
g = 9.8
L = 1
hasta_tiempo = 5
dt = 0.0001
angulo_inicial = np.pi / 18
v_angular_inicial = 0

w = np.sqrt(g/L)
amplitud = np.sqrt(angulo_inicial**2 + (v_angular_inicial / w)**2)
fase = np.arcsin(angulo_inicial / amplitud)

t = np.arange(0, hasta_tiempo, dt)

angulo_t = amplitud * np.sin(w*t + fase)

plt.title('Solucion Analitica')
plt.ylabel('Amplitud [rad⁻¹]')
plt.xlabel('tiempo [s]')
plt.xticks(np.arange(0, hasta_tiempo + 1,  2))
plt.plot(t, angulo_t, 'navy')

plt.show()
```

# SOLUCIÓN NUMÉRICA

```python
tiempos = []
angulos = []

def  deltas(dt, angulo, velocidad): 
  global g, L
  dvel_dt = - (g / L) * np.sin(angulo) * dt
  dang_dt= velocidad * dt
  return dvel_dt, dang_dt

angulo = angulo_inicial
velocidad = v_angular_inicial
for i in t:
  dvel, dang = deltas(dt, angulo, velocidad)
  velocidad += dvel
  angulo += dang

  tiempos.append(i)
  angulos.append(angulo)

plt.title('Solucion Numerica')
plt.xticks(np.arange(0, hasta_tiempo + 1,  2))
plt.plot(tiempos, angulos, 'red')
plt.show()
```

# COMPARACIÓN


```python
plt.plot(tiempos, angulos, color = 'tab:red', lw = 5, alpha = 0.5, label = 'numerica')
plt.plot(t, angulo_t, 'black', lw = 2, label = 'analitica')
plt.legend()
plt.show()
```
